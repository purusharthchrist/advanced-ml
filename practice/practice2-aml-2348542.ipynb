```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data_power_transformed.csv')
df.head()
```

```python
#dropping unnecessary attributes
df.drop(['Unnamed: 0', 'date', 'street', 'statezip', 'country'], axis=1, inplace=True)
df.head()
```

```python
#value counts for 'city' attribute
df.city = [dict(df.city.value_counts())[_] for _ in df.city]
df.head()
```

```python
from scipy.stats import zscore

df.city = zscore(df.city)
df.head()
```

# Splitting the data

```python
from sklearn.model_selection import train_test_split

X, y = df.drop('price', axis=1).values, df.price.values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# KNN

```python
from sklearn.neighbors import KNeighborsRegressor

knr = KNeighborsRegressor()
knr.fit(X_train, y_train)
```

```python
#evaluating the model
from sklearn.metrics import accuracy_score

y_pred = knr.predict(X_test)

knr.score(X_test, y_test)
```
